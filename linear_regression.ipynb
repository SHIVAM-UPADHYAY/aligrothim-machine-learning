```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv('headbrain.csv')
data.head()
X=data['Head Size(cm^3)'].values
Y=data['Brain Weight(grams)'].values

mean_x=np.mean(X)
mean_y=np.mean(Y)

n=len(X)
numerator=0
denominator=0

for i in range (n):
    numerator+=(X[i]-mean_x)*(Y[i]-mean_y)
    denominator+=(X[i]-mean_x)**2
    
b1=numerator/denominator
b0=mean_y-(b1*mean_x)

print(b1)
print(b0)

min_x=np.min(X)-100
max_x=np.max(X)+100

x=np.linspace(min_x,max_x,1000)
y=b0+b1*x

plt.plot(x,y)
plt.scatter(X,Y,color='#EF5423')
plt.show()


```

```python
#calculating R2 Score
ss_total=0
ss_res=0

for i in range (n):
    y_pred=b0+b1*X[i]
    ss_total+=(Y[i]-np.mean(Y))**2
    ss_res+=(Y[i]-y_pred)**2
    
r2=1-(ss_res/ss_total)
print("R2 Score ")
print(r2)
```
